```python
# LIBROS
class libros:
    def __init__(self,serie, titulo, autor):
        self.serie = serie
        self.titulo = titulo
        self.autor = autor
        self.disponible = True

    def prestamo(self):
        if self.disponible == True:
            self.disponible = False
            print("El libro '",self.titulo,"' estaba disponible. ¬°Disfruta tu lectura!")
        else:
            print("El libro '",self.titulo,"' no est√° disponible. De  momento ya ha sido prestado.")

    def devolucion(self):
        if self.disponible == False:
            self.disponible = True
            print("Gracias por devolver el libro '",self.titulo,"' ! Nos vemos pronto")
        else:
            print("Gracias por tu visita. Pero ese libro '",self.titulo,"' ya lo tenemos")

libro1 = libros(1,"Cien a√±os de soledad", "Gabriel Garc√≠a M√°rquez")
libro2 = libros(2,"Don Quijote de la Mancha", "Miguel de Cervantes")
libro3 = libros(3,"Rayuela", "Julio Cort√°zar")
libro4 = libros(4,"Pedro P√°ramo", "Juan Rulfo")
libro5 = libros(5,"La sombra del viento", "Carlos Ruiz Zaf√≥n")
libro6 = libros(6,"El amor en los tiempos del c√≥lera", "Gabriel Garc√≠a M√°rquez")
libro7 = libros(7,"Ficciones", "Jorge Luis Borges")
libro8 = libros(8,"La casa de los esp√≠ritus", "Isabel Allende")
libro9 = libros(9,"Como agua para chocolate", "Laura Esquivel")
libro10 = libros(10,"El t√∫nel", "Ernesto Sabato")

print("Libros disponibles:")
libros_ejemplo = [libro1, libro2, libro3, libro4, libro5, libro6, libro7, libro8, libro9, libro10]
for libro in libros_ejemplo:
    print(f"Serie: {libro.serie}, T√≠tulo: {libro.titulo}, Autor: {libro.autor}, Disponible: {'S√≠' if libro.disponible else 'No'} \n ")


#USUARIO


libro1.prestamo()
libro1.prestamo()
libro1.devolucion()
print(libro1.serie)

```

```python
from typing import Any


class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True
    
    def borrow(self):
        if self.available:
            self.available = False
            print(f"El libro {self.title} ha sido prestado")
        else:
            print(f"El libro {self.title} no est√É¬° disponible")

    def return_book(self):
        self.available = True
        print(f"El libro {self.title} ha sido devuelto")

class User:
    def __init__(self, name, user_id):
        self.name = name
        self.user_id = user_id
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.available:
            book.borrow()
            self.borrowed_books.append(book)
        else:
            print(f"El libro {book.title} No esta disponible")

    def return_book(self, book):
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
        else:
            print(f"El libro {book.title} No esta en la lista de prestados")

class Library:
    def __init__(self):
        self.books = []
        self.users = []
    
    def add_book(self, book):
        self.books.append(book)
        print(f"El libro {book.title} ha sido agregado")

    def register_user(self, user):
        self.users.append(user)
        print(f"El usuario {user.name} ha sido registrado")

    def show_available_books(self):
        print("Libros disponibles:")
        for book in self.books:
            if book.available:
                print(f"{book.title} por {book. author}")

#Crear los libros
book1 = Book("El principito", "Antoine de Saint-Exup√É¬©ry")
book2 = Book("1984", "George Orwell")

#Crear usuario
user1 = User("Carli", "001")

#Crear Biblioteca
library = Library()
library.add_book(book1)
library.add_book(book2)
library.register_user(user1)

#Mostrar libros
library.show_available_books()

#Realizar prestamo
user1.borrow_book(book1)

#Mostrar libros
library.show_available_books()

#Devolver libro
user1.return_book(book1)

#Mostrar libros
library.show_available_books()
```
