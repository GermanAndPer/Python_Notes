#  Librerias y como importarlas 

___Qu es una Librera y qu es un Mdulo?___

- `Librera`: En Python, una librera es un conjunto organizado de mdulos que puedes usar para aadir funcionalidades a tu cdigo sin tener que escribirlas t mismo. Piensa en ello como una coleccin de herramientas especializadas listas para usar.

- `Mdulo`: 
Un mdulo es un archivo de Python que contiene cdigo que puedes reutilizar en tus proyectos. Un mdulo puede incluir funciones, clases, y variables que te ayudan a resolver problemas especficos de manera eficiente.



Hay librerias predeterminadas que ayudan a la muchas cosas. Como:
- `math`: Para realizar operaciones matemticas.
- `random`: Para generar nmeros aleatorios.
- `time`: Para trabajar con fechas y horas.
- `os`: Para interactuar con el sistema operativo.
- `json`: Para trabajar con datos en formato JSON.
- `re`: Para trabajar con expresiones regulares.
- `datetime`: Para trabajar con fechas y horas.
- `csv`: Para trabajar con archivos CSV.
- `pandas`: Para trabajar con datos en forma de tablas.
- `numpy`: Para trabajar con matrices y vectores numricos.
- `matplotlib`: Para crear grficos.
- `seaborn`: Para crear grficos de alta calidad.
- `scikit-learn`: Para realizar operaciones de aprendizaje automtico.
- `scipy`: Para realizar operaciones cientficas.
- `tkinter`: Para crear interfaces grficas de usuario.
- `pygame`: Para crear juegos.
- `requests`: Para realizar solicitudes HTTP.
- `urllib`: Para realizar operaciones con URLs.
- `xml.etree.ElementTree`: Para trabajar con archivos XML.
- `xml.dom`: Para trabajar con documentos XML.


Si los dividimos por reas se pueden clasificar as.

- `Manejo de Archivos y Sistema`: Mdulos como ___os___, ___shutil___, y ___pathlib___ te permiten interactuar con el sistema de archivos, lo cual es esencial para casi cualquier proyecto.
- `Operaciones Matemticas: Mdulos como ___math___ y ___random___ te proporcionan funciones matemticas avanzadas y generacin de nmeros aleatorios.
- `Manejo de Fechas y Tiempos`: ___datetime___ y ___time___ te permiten trabajar con fechas y horas, lo cual es crucial para la programacin de eventos o el registro de actividades.
- `Manipulacin de Datos`: Mdulos como ___json___ y ___csv___ son ideales para leer y escribir datos estructurados, algo comn en el manejo de APIs y almacenamiento de informacin.
- `Redes y Comunicaciones`: Si ests construyendo aplicaciones que necesitan comunicarse a travs de una red, ___socket___ y ___http.server___ te proporcionan las herramientas necesarias para gestionar conexiones y servidores web.
