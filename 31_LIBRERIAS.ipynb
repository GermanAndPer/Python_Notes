# 📚 Librerias y como importarlas 

___¿Qué es una Librería y qué es un Módulo?___

- `Librería`: En Python, una librería es un conjunto organizado de módulos que puedes usar para añadir funcionalidades a tu código sin tener que escribirlas tú mismo. Piensa en ello como una colección de herramientas especializadas listas para usar.

- `Módulo`: 
Un módulo es un archivo de Python que contiene código que puedes reutilizar en tus proyectos. Un módulo puede incluir funciones, clases, y variables que te ayudan a resolver problemas específicos de manera eficiente.



Hay librerias predeterminadas que ayudan a la muchas cosas. Como:
- `math`: Para realizar operaciones matemáticas.
- `random`: Para generar números aleatorios.
- `time`: Para trabajar con fechas y horas.
- `os`: Para interactuar con el sistema operativo.
- `json`: Para trabajar con datos en formato JSON.
- `re`: Para trabajar con expresiones regulares.
- `datetime`: Para trabajar con fechas y horas.
- `csv`: Para trabajar con archivos CSV.
- `pandas`: Para trabajar con datos en forma de tablas.
- `numpy`: Para trabajar con matrices y vectores numéricos.
- `matplotlib`: Para crear gráficos.
- `seaborn`: Para crear gráficos de alta calidad.
- `scikit-learn`: Para realizar operaciones de aprendizaje automático.
- `scipy`: Para realizar operaciones científicas.
- `tkinter`: Para crear interfaces gráficas de usuario.
- `pygame`: Para crear juegos.
- `requests`: Para realizar solicitudes HTTP.
- `urllib`: Para realizar operaciones con URLs.
- `xml.etree.ElementTree`: Para trabajar con archivos XML.
- `xml.dom`: Para trabajar con documentos XML.


Si los dividimos por áreas se pueden clasificar así.

- `Manejo de Archivos y Sistema`: Módulos como ___os___, ___shutil___, y ___pathlib___ te permiten interactuar con el sistema de archivos, lo cual es esencial para casi cualquier proyecto.
- `Operaciones Matemáticas: Módulos como ___math___ y ___random___ te proporcionan funciones matemáticas avanzadas y generación de números aleatorios.
- `Manejo de Fechas y Tiempos`: ___datetime___ y ___time___ te permiten trabajar con fechas y horas, lo cual es crucial para la programación de eventos o el registro de actividades.
- `Manipulación de Datos`: Módulos como ___json___ y ___csv___ son ideales para leer y escribir datos estructurados, algo común en el manejo de APIs y almacenamiento de información.
- `Redes y Comunicaciones`: Si estás construyendo aplicaciones que necesitan comunicarse a través de una red, ___socket___ y ___http.server___ te proporcionan las herramientas necesarias para gestionar conexiones y servidores web.

# ⬇️ Como instalarlas

## ✅ Ya integradas

Si la libreria ya esta integrada, simplemente utilizamos IMPORT al inicio de nuestro archivo de Python

```python
#EJEMPLO
import random
```

## ⬇️Para Instalar

Para instalar nuevas librerias se necesita de la linea de comandos. utilizamos la palabra reservada `PIP`

___PIP___ nos ayuda a instalar librerias desde origenes externos.


```python
#Ejemplo génerico
!pip install nombre-de-la-libreria

#En algunos casos no permite usar el ! por lo que se sugiere usar el %
%pip install nombre-de-la-libreria

#Ejemplo practico
!pip install pandas
%pip install pandas

#NOTA: Se usa ! o % para acceder a la linea de comandos desde python. en caso de aplicarla directamente sobre la linea de comandos, no se necesitará.
```
