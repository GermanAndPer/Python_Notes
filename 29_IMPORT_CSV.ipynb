# üßÆ CSV

Python tiene una libreria que podemos importar llamada CSV.

La cual fasilita a la lectura y analisis de archivos  CSV. El formato mas comun para exportar Datos.

Estos son los usos mas comunes:
- `reader()` : Lee el archivo CSV y devuelve un objeto de lectura que contiene una lista de listas, donde cada lista representa una fila del archivo CSV.
- `writer()` : Crea un objeto de escritura que permite escribir en archivos CSV.
- `writerow()` : Escribe una fila en el archivo CSV.
- `writerows()` : Escribe varias filas en el archivo CSV.
- `DictReader()` : Lee el archivo CSV y devuelve un objeto de lectura que contiene un diccionario para cada fila del archivo CSV.
- `DictWriter()` : Crea un objeto de escritura que permite escribir en archivos CSV con un diccionario para cada fila.
- `register_dialect()` : Registra un dialecto de CSV personalizado.
- `get_dialect()` : Obtiene un dialecto de CSV registrado.
- `list_dialects()` : Obtiene una lista de dialectos de CSV registrados
- `SNiffer()` : Detecta el dialecto de un archivo CSV.

```python
import csv

# DICT_READER
# Dictreader nos ayuda a leer un archivo CSV como diccionario guardando esos datos dentro de una variable
# CSV
# Cuando creamos la variable y asignamos el valor, usamos CSV para decir que va a leer un CSV
# DictReader
# Es un objeto que nos permite leer un archivo CSV y convertirlo en un diccionario
# Es importante mencionar que el archivo CSV debe tener LOS ENCABEZADOS EN La PRIMER LINEA
# DATOS
# En este caso se usa la variable DATOS por que es donde se estan guardando los valores del archivo.
# Pero pueden ser cualquier nombre. Siempre que se el que hace referencia a los datos del CSV

# CICLO FOR
# El CICLO FOR es para comprobar que se haya aplicado correctamente. 

with open("Data.csv","r") as datos:
    lectura_como_diccionario = csv.DictReader(datos)
    conteo = 0
    for fila in lectura_como_diccionario:
        print(fila)
        conteo += 1
        #print(fila["id"])
    print(conteo)
```

```python
import csv
#NUEVO REGISTRO

nueva_linea = {
    'id': '1001', 
    'first_name': 'Clary', 
    'last_name': 'Maylour', 
    'email': 'cmaylour0@artisteer.com', 
    'gender': 'Female', 
    'ip_address': '239.137.134.155'
    }

# DICT_WRITER
# Para escribir una nueva linea utilizamos el metodo DICTWRITER y el modo de lectura A
# MODE=A
# Aunque para el modo de lectura podemos poner simplemente A tambien podemos ser mas presiso y escribir MODE=A
# NEWLINE=""
# Se recomienda usar newline="" para evitar problemas con los saltos de l√≠nea. pues dependiendo del sistema pueden duplicarse los saltos de linea.
# Al ponerlo textualmente, Python no modifica los saltos de l√≠nea y permite que el m√≥dulo csv los maneje correctamente.
# DICT_WRITER
# Tiene una estructura diferente.
# Primero pide los el archivo al que se va a gregar  la informacion. Que en este caso es DATOS
# Despues solicita los encabezados del archivo que vamos a importar. Podemos usar fieldnames=nueva_linea.keys() para automatizar.
# FIELDNAME
# Es para especificar que vas a dar los nombres de los campos.
# =nueva_linea.keys()
# Es para sacar las claves del nuevo registro. 
# Pero tambien pueden darse todos las claves de manera manual.
# WRITEROW
# Es para escribir una nueva linea en el archivo.

with open("Data.csv",mode="a",newline="") as datos:
    escribir_como_diccionario = csv.DictWriter(datos, fieldnames=nueva_linea.keys())   
    #escribir_como_diccionario = csv.DictWriter(datos, fieldnames= ['id', 'first_name', 'last name', 'email', 'gender', 'ip_address'])
    escribir_como_diccionario.writerow(nueva_linea)
                                                 # CERRAR ARCHIVO
```

```python
# NUEVA COLUMNA
# Empezamos leyendo el archivo original.
# una vez que lo leimos, agregamos el nuevo campo que queramos agregar con .FILENAMES
# Y le agredamos el nuevo encabezado como lista
# csv_reader.fieldnames + ['NOMBRE DE NUEVA COLUMNA'] 
# Luego lo escribimos en un nuevo archivo
# El nombre que le demos, si no existe va a generar uno. Y ahora se habre con el modo W. Por que se va a escribir de 0
# WRITEHEADER
# Una vez creados e indicados los encabezados, confirmamos que la primera fila son los encabezados.

# CICLO FOR
# Recorremos cada fila del archivo original para llenar el nuevo archivo.
# Pero para cada ROW asignamos el nuevo valor que con la operacion que se necesite

with open("Data.csv","r") as datos:
    lectura_como_diccionario = csv.DictReader(datos)
    nombre_de_campos = lectura_como_diccionario.fieldnames + ["Full Name"]

    with open("Data_v2.csv",mode="w",newline="") as datos_leidos:
        nuevo_archivo = csv.DictWriter(datos_leidos, fieldnames=nombre_de_campos)
        nuevo_archivo.writeheader()

        for linea in lectura_como_diccionario:
            linea["Full Name"] = linea["first_name"] + " " + linea["last_name"]
            nuevo_archivo.writerow(linea)  # Agregamos la nueva fila al archivo nuevo
                                                                   
```
