# 🧮Matrices 

Las matrices son listas con listas.

Ejemplo de 3 listas:



```python
# Matrices
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(matrix)
print(matrix[0][1])  # Output: 2
print(matrix[1][2])  # Output: 6
# Tuplas
# Las tuplas son similares a las listas, pero son inmutables.
# Esto significa que una vez que se crea una tupla, no se puede modificar.
# Se definen utilizando paréntesis en lugar de corchetes.
tuple_example = (1, 2, 3)
print(tuple_example[0])  # Output: 1
# Las tuplas son útiles cuando se necesita una colección de elementos que no cambiarán.
# Por ejemplo, se pueden usar para almacenar coordenadas o datos que no deben modificarse.
# Las tuplas también pueden contener diferentes tipos de datos, al igual que las listas.
# Sin embargo, a diferencia de las listas, no se pueden agregar, eliminar o modificar elementos
# dentro de una tupla una vez que se ha creado.
# Esto las hace más eficientes en términos de memoria y rendimiento en ciertas situaciones.
# Las tuplas se pueden utilizar en situaciones donde se requiere una colección de datos
# que no cambiará a lo largo del tiempo, como coordenadas geográficas, fechas o
# configuraciones que no deben ser alteradas.

#La misma TUPLA se puede crear sin paréntesis, separando los valores por comas.
tuple_example2 = 4, 5, 6
print(tuple_example2[1])  # Output: 5
```

Como recomendacion a las MATRICES, mientras cada vez esten anidadas, se recomienda ordenarlas para identificar sus posiciones.
