```python
# Lista normal
# Lista de enteros
lista_de_numeros = [1, 2, 3, 4, 5]
print("Tu lista de numeros son:",lista_de_numeros)

#Lista mixta
##Una lista puede contener diferentes tipos de datos
#Por ejemplo, una lista puede contener enteros, cadenas de texto, booleanos, etc.
lista_mixta = [1, "Hola", 3.14, True]
print("Tu lista mixta es:",lista_mixta)

#Lista vacia
#Una lista vacia es una lista que no contiene ningun elemento
lista_vacia = []
print("Tu lista vacia es:",lista_vacia)

#Lista de String
#Una cadena de texto es un listado de caracteres. Por lo tanto, una lista de String es una lista que contiene solo cadenas de texto.
lista_string = "Hola, Mundo"
print("Tu lista de String es:", lista_string)

#METODOS
#Los metodos son funciones que se aplican a un objeto. En este caso, los metodos se aplican a una lista.

#LEN
#Devuelve el numero de elementos en una lista.
print("Tu lista de numeros tiene ",len(numeros),"elementos.")

#APPEND
#Agregar un elemento al final de una lista
lista_de_numeros.append(6)
print("Tu lista de numeros ahora es:", lista_de_numeros)

#EXTEND
#Agregar varios elementos al final de una lista
#O agregar una lista a otra lista
lista_de_numeros.extend([7, 8, 9])
print("Tu lista de numeros ahora es:", lista_de_numeros)

#INSERT
#Agregar un elemento una lista. Indicando la posici√≥n donde se va a agregar el elemento.
#En este caso, se agrega el elemento 0 en la posici√≥n 1 de la lista.
#Las posiciones de una lista empiezan desde 0, por lo que la posici√≥n 1 es el segundo elemento de la lista.
#Si la posici√≥n no existe, se agrega al final de la lista.
#Si la posici√≥n es negativa, se cuenta desde el final de la lista.
lista_de_numeros.insert(9, 10)
print("Tu lista de numeros ahora es:", lista_de_numeros)

#POP
#Eliminar el √∫ltimo elemento de una lista
#El m√©todo pop elimina el √∫ltimo elemento de una lista y lo devuelve.
#Si se quiere eliminar un elemento en una posici√≥n espec√≠fica, se puede indicar la posici√≥n como argumento.
#Si la posici√≥n no existe o est√° vac√≠a, se genera un error.
#El elemento eliminado se puede almacenar en una variable para poder utilizarlo m√°s adelante. (Es opcional)
ultimo_elemento = lista_de_numeros.pop()
print("El √∫ltimo elemento eliminado es:", ultimo_elemento)
print("Tu lista de numeros ahora es:", lista_de_numeros)

#LIST
#Para ver una posici√≥n espec√≠fica de una lista, se puede utilizar el √≠ndice de la posici√≥n.
print("El primer elemento de la lista es:", lista_de_numeros[0])
print("El segundo elemento de la lista es:", lista_de_numeros[1])

#INDEX
#Para ver una posici√≥n espec√≠fica de una lista, se puede utilizar el √≠ndice de la posici√≥n.
#El m√©todo index devuelve el √≠ndice del primer elemento que coincide con el valor indicado.
#Si el valor no se encuentra en la lista, se genera un error.
print("El √≠ndice del elemento 3 es:", lista_de_numeros.index(3))
print("El √≠ndice del elemento 9 es:", lista_de_numeros.index(9))

#MAX
#El elemento m√°s grande de una lista
print("El elemento m√°s grande de la lista es:", max(lista_de_numeros))

#MIN
#El elemento m√°s peque√±o de una lista
print("El elemento m√°s peque√±o de la lista es:", min(lista_de_numeros))

#COUNT
#Contar el n√∫mero de veces que un elemento aparece en una lista
#El m√©todo count devuelve el n√∫mero de veces que un elemento aparece en una lista.
print("El n√∫mero de veces que el elemento 3 aparece en la lista es:", lista_de_numeros.count(3))

#IN
#Verificar si un elemento est√° en una lista
#El m√©todo in devuelve True si el elemento est√° en la lista, False en caso contrario.
print("El elemento 3 est√° en la lista:", 3 in lista_de_numeros)
print("El elemento 10 est√° en la lista:", 10 in lista_de_numeros)

#REVERSE
#Invertir el orden de los elementos en una lista
#El m√©todo reverse invierte el orden de los elementos en una lista.
lista_de_numeros.reverse()
print("Tu lista de numeros despues de REVERSE es:", lista_de_numeros)

# SORT
# Ordenar los elementos de una lista
# El m√©todo sort ordena los elementos de una lista en orden ascendente.
lista_de_numeros.sort()
print("Tu lista de numeros despues de SORT es:", lista_de_numeros)

#VALUEERROR
#Error si el elemento no existe
#Si se intenta acceder a un elemento que no existe en la lista, se genera un error
#Por ejemplo, si se intenta acceder o eliminar la posici√≥n 10 de una lista que tiene solo 5 elementos, se genera un error.#Esto tambi√©n ocurre si se intenta eliminar un elemento que no existe en la lista.
print("Intentando acceder a un elemento que no existe en la lista:")
#TRY
#El bloque try permite manejar errores de forma controlada.
#Si se genera un error, se ejecuta el bloque except.
#En este caso, se intenta acceder a un elemento que no existe en la lista.
#Si se genera un error, se imprime un mensaje de error.
#Siendo e un objeto que contiene informaci√≥n sobre el error.
#El error generado es del tipo IndexError. Indica que se est√° intentando acceder a un √≠ndice que no existe en la lista.
try:
    print(lista_de_numeros[10])
except IndexError as e:
    print("Error:", e)


#REMOVE
#Eliminar un elemento de una lista por su valor
#El m√©todo remove elimina el primer elemento que coincide con el valor indicado.
lista_de_numeros.remove(4)
print("Tu lista de numeros despues de REMOVE es:", lista_de_numeros)

#DEL - Eliminar un elemento de una lista por su √≠ndice
#Eliminar un elemento de una lista por su √≠ndice
del lista_de_numeros[2]
print("Tu lista de numeros despues de DEL es:", lista_de_numeros)
#El m√©todo del elimina el elemento en la posici√≥n indicada.
#Si la posici√≥n no existe, se genera un error.
#En este caso, se elimina el tercer elemento de la lista (√≠ndice 2).
#Si se intenta eliminar un elemento que no existe en la lista, se genera un error.

#CLEAR
#Eliminar todos los elementos de una lista
#El m√©todo clear elimina todos los elementos de una lista.
lista_de_numeros.clear()
print("Tu lista de numeros despues de CLEAR es:", lista_de_numeros)

# COPY
# Crear una copia de una lista
# El m√©todo copy crea una copia de una lista.
lista_de_numeros = [1, 2, 3, 4, 5]
copia_lista = lista_de_numeros.copy()
print("Tu lista de numeros es:", lista_de_numeros)
print("Tu copia de la lista de numeros es:", copia_lista)

#DEL - Eliminar una lista
#Eliminar una lista
#El m√©todo del elimina una lista.
del lista_de_numeros
try:
    print("Tu lista de numeros es:", lista_de_numeros)
except NameError as e:
    print("Error:", e)
```
